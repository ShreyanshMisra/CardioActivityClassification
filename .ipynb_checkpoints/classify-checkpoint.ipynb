# Classifying Cardio Activities Using Sensor Data

Health Tracking devices and algorithms aren’t always optimized for gym cardio equipment, even though most people do a lot of their daily physical activity at the gym. We conducted an activity classification investigation—classifying the three most popular cardio activities at the UMass recreation center: stairmaster, incline treadmill, and stationary bike—then comparing the effectiveness of a accelerometer approach to a accelerometer and gyroscope approach.

## Data Preperation

### Import Dependencies

```python
import pandas as pd
import glob
import re
import os
import sys
import pickle
import datetime
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn import tree, metrics
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
from scipy.signal import butter, filtfilt, find_peaks
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn.model_selection import train_test_split
```

### Helper Functions

```python
def calc_magnitude(data):

    data['accel_mag'] = np.sqrt(data['x']**2 + data['y']**2 + data['z']**2) # absolute accel magnitude
    data['accel_mag'] = data['accel_mag'] - data['accel_mag'].mean() # detrend: "remove gravity"

    return data
```

```python
def remove_noise(data,sampling_rate):
    from scipy.signal import butter, filtfilt, find_peaks

    # Low pass filter
    cutoff = 5 # Hz
    order = 2
    b, a = butter(order, cutoff/(sampling_rate/2), btype='lowpass')
    data['filtered_accel_mag'] = filtfilt(b, a, data['accel_mag'])

    return data
```

```python
def add_features(window):
    features = {}
    features['avg'] = window['filtered_accel_mag'].mean()
    features['max'] = window['filtered_accel_mag'].quantile(1)
    features['med'] = window['filtered_accel_mag'].quantile(0.5)
    features['min'] = window['filtered_accel_mag'].quantile(0)
    features['q25'] = window['filtered_accel_mag'].quantile(0.25)
    features['q75'] = window['filtered_accel_mag'].quantile(0.75)
    features['std'] = window['filtered_accel_mag'].std()
    df = pd.DataFrame()
    df = df._append(features,ignore_index=True)
    return df
```

```python
def train_decision_tree(frames):
    # Extract feature columns
    X = frames[['avg', 'max', 'med', 'min', 'q25', 'q75', 'std']]

    # Extract target column
    y = frames['activity']

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Create model
    dt_model = DecisionTreeClassifier(criterion='entropy',max_depth=5).fit(X_train, y_train)
    dt_pred = dt_model.predict(X_test)

    # Evaluate on test set
    acc = dt_model.score(X_test, y_test)
    dt_cm = confusion_matrix(y_test, dt_pred, labels=dt_model.classes_)
    print(classification_report(y_test, dt_pred))
    print("Accuracy on test set:", acc)

    return dt_model,dt_cm,acc
```

```python
def train_decision_tree_combined(frames):
    # Extract feature columns
    X = frames[['accel_avg', 'accel_max', 'accel_med', 'accel_min', 'accel_q25', 'accel_q75', 'accel_std',
                                     'gyro_avg', 'gyro_max', 'gyro_med', 'gyro_min', 'gyro_q25', 'gyro_q75', 'gyro_std']]

    # Extract target column
    y = frames['activity']

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Create model
    dt_model = DecisionTreeClassifier(criterion='entropy',max_depth=5).fit(X_train, y_train)
    dt_pred = dt_model.predict(X_test)

    # Evaluate on test set
    acc = dt_model.score(X_test, y_test)
    dt_cm = confusion_matrix(y_test, dt_pred, labels=dt_model.classes_)
    print(classification_report(y_test, dt_pred))
    print("Accuracy on test set:", acc)

    return dt_model,dt_cm,acc
```

```python
# Function to extract windows and features
def extract_features(data, window_sec, sample_rate, activity):
    
    all_features = []
    
    for timestamp, window in data.resample(f'{window_sec}S'):
        features = add_features(window)
        features['activity'] = activity
        
        all_features.append(features)
    
    results = pd.concat(all_features, ignore_index=True)
    
    return results
```

```python
def all_data_to_combined_csv(root, output_filename = 'all_data.csv'):
    
    merged = pd.DataFrame()
    
    activities = ['stairmaster', 'bike', 'treadmill']
    
    for activity in activities:
        folder = os.path.join(root, activity)
        raw_files = glob.glob(os.path.join(folder, '*.csv'))
        
        for file in raw_files:
            data = pd.read_csv(file, parse_dates=['time'],index_col='time')
            data = calc_magnitude(data)
            data = remove_noise(data, sampling_rate=100)
            
            features = extract_features(data, window_sec=10, sample_rate=100, activity=activity)
            merged = pd.concat([merged, features], ignore_index=True)
            
    merged.to_csv(os.path.join(root, output_filename), index=False)
```

```python
def transform_time_to_datetime(root):
    
    dateparse = lambda dates: [pd.to_datetime(d, unit='ns') for d in dates]
    # Get list of all activity folders
    activity_folders = os.listdir(root)
    # print(activity_folders)

    for folder in activity_folders:
        # print(folder)
        files = glob.glob(f"{root}/{folder}/*.csv")
        for filename in files:
            # print(filename)
            df = pd.read_csv(filename, parse_dates=['time'])
            df['time'] = pd.to_datetime(pd.to_numeric(df['time']), unit='ns')
            df.to_csv(filename, index=False)
```

```python
# Helper function to visualize model - Do not modify
def viz_tree(dt_model,features_frames,cnames):
    # Fix feature names as list
    feature_names = features_frames.columns.tolist()

    fig, ax = plt.subplots(figsize=(9,4))
    tree.plot_tree(dt_model,
                   feature_names=feature_names,
                   fontsize=7,
                   class_names=cnames,
                   filled=True,
                   ax=ax)

    plt.title('Decision Tree')
    plt.savefig('dt.png')
```

### Combine Data

```python
def all_data_to_combined_csv(root, output_filename = "all_files.csv"):
    
    merged = pd.DataFrame()
    
    activities = ['Bike', 'Stairmaster', 'Treadmill']
    
    for activity in activities:
        folder = os.path.join(root, activity)
        raw_files = glob.glob(os.path.join(folder, '*.csv'))
        
        for file in raw_files:
            data = pd.read_csv(file, parse_dates=['time'],index_col='time')
            data = calc_magnitude(data)
            data = remove_noise(data, sampling_rate=100)
            
            features = extract_features(data, window_sec=10, sample_rate=100, activity=activity)
            merged = pd.concat([merged, features], ignore_index=True)
            
    merged.to_csv(os.path.join(root, output_filename), index=False)
```

```python
accel = pd.read_csv('./data/Accelerometer/all_files.csv')
gyro = pd.read_csv('./data/Gyroscope/all_files.csv')
```

```python
activities = ['Stairmaster','Bike','Treadmill']

filtered_accel = accel[accel['activity'].isin(activities)]
filtered_gyro = gyro[gyro['activity'].isin(activities)]

len(filtered_accel), len(filtered_gyro)
```

### Accelerometer Only

```python
dt_model, dt_cm, acc = train_decision_tree(filtered_accel)

with open(f'dt_model-{str(activities)}.pkl', 'wb') as f:
    pickle.dump(dt_model, f)
    
cm_display = ConfusionMatrixDisplay(confusion_matrix = dt_cm, display_labels=dt_model.classes_)
cm_display.plot()
plt.show()

viz_tree(dt_model,filtered_accel,filtered_accel['activity'].unique().tolist())
```

```python
importances = dt_model.feature_importances_

feature_names = ['avg', 'max', 'med', 'min', 'q25', 'q75', 'std']
importance_dict = dict(zip(feature_names, importances))
sorted_importances = sorted(importance_dict.items(), key=lambda x: x[1], reverse=True)
print("Sorted feature importances:", sorted_importances)


print("Feature importances:", importances)
```
